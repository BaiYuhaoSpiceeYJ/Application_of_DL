```python
import os
import subprocess
import pickle
import glob
from music21 import converter, instrument, note, chord, stream#note音符，chord和弦

def get_notes():
    
    
    
    stream = converter.parse(".\\AI_make_music\\1.mid")#解析文件

    
    parts = instrument.partitionByInstrument(stream)#获取所有乐器部分。

    if parts:  # 如果有乐器部分， 取第一个乐器部分，为钢琴
        notes = parts.parts[0].recurse()
    else: #否则为一个纯的音符组成
        notes = stream.flat.notes

    for element in notes:# 如果是 Note 类型，那么取它的音调
        print(str(element))
        

    return notes

get_notes()
```
